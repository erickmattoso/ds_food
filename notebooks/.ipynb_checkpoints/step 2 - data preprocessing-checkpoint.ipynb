```python
import pandas as pd
import re
```

```python
pd.set_option('display.max_colwidth', 100)
pd.set_option("display.max_rows",    1000)
data = pd.read_csv('data/data_aug_5.csv',index_col=0)
```

```python
def remove_new_line(char0):
    return data.replace(char0,'', regex=True, inplace= True)

def remove_char(char0):
    return data['caracteristicas'].replace(char0,'', regex=True, inplace= True)

def remove_comma():
    data['caracteristicas'].replace("'", "",           regex=True, inplace= True)
    data['caracteristicas'].replace(' +', ' ',         regex=True, inplace= True)
    data['caracteristicas'].replace(' ,', '',          regex=True, inplace= True)
    data['caracteristicas'].replace(',', "','",        regex=True, inplace= True)
    data['caracteristicas'].replace("\[',' ", "['",    regex=True, inplace= True)
    data['caracteristicas'].replace("',' ]| ]", "']",  regex=True, inplace= True)
    data['caracteristicas'].replace(",' ", ",'",       regex=True, inplace= True)

remove_new_line(r"""\\n""")    
remove_char(r"""<b>|</b>|<br/>|</span>|Grau de dificuldade:|Tempo de preparo:|Rendimento:|
|<span itemprop="recipeYield">|
|<meta content="PT30M" itemprop="prepTime"/>|
|<meta content="PT1H" itemprop="prepTime"/>""")
remove_char(r""" <span itemprop="nutrition" itemscope="" |
|itemtype="http://schema.org/NutritionInformation">Valor nutricional: <span itemprop="calories">""")
remove_comma()
```

```python
pd.options.mode.chained_assignment = None  # default='warn'
data['caracteristicas'].iloc[1599] = """['Fácil','Até meia hora','2 pães']"""
data['caracteristicas'].iloc[2187] = """['Fácil','Até meia hora']"""
data['caracteristicas'].iloc[2427] = """['Fácil','Até meia hora']"""
data['caracteristicas'].iloc[2467] = """['Fácil','Até meia hora']"""
data['caracteristicas'].iloc[356]  = """['Médio','Até uma hora','1.5 kg']"""
data['caracteristicas'].iloc[883]  = """['Fácil','Até uma hora','1.2 kg ou 20 pedaços']"""
data['caracteristicas'].iloc[999]  = """['Fácil','Até meia hora','1.5 kg de torta']"""
data['caracteristicas'].iloc[1998] = """['Fácil','Até meia hora','para perus de 3 kg/3.5 kg', '339 calorias (porção) ou 2.034 (receita)']"""
```

```python
filter_1 = data['caracteristicas'].str.contains('porções[0-9]+')
data.loc[filter_1, 'caracteristicas'] = data[filter_1]['caracteristicas']\
    .replace("porções", "porções','", regex=True,)
```

```python
data.to_csv('data/data_clr.csv')
data.info()
```
