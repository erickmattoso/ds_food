```python
dasd = []
for tag in soup.find("div", { "class" : "recipe_image" }).next_siblings:
    if soup.find("div", { "itemprop" : "description" }) == tag:
        break
    else:
        dasd.append(tag)
dasd
```

```python
url = 'http://receitas.folha.com.br/receita/2576'
req = requests.get(url)
soup = BeautifulSoup(req.text, "lxml")
```

```python
imgs = soup.findAll("img", {"itemprop":"image"})
# download img
for img in imgs:
#     # get the src of the img
    img_url = urljoin(url, img['src'])
#     # it split all src link and get the final part to create the name of the file
    file_name = img['src'].split('/')[-1]
#     # save this file on a folder called "img/"
    file_path = os.path.join("img/", file_name)
#     # actually downloads the img
    urlretrieve(img_url, file_path)


file_path
```
