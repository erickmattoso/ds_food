```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from urllib.parse import urljoin
from urllib.request import urlretrieve
```

```python
#funcao para crawlear o site receitas.folha.com.br
def request_receipt():
    print('começou \n')
    page = []

    for i in range(2960, -1, -1):
        tag_title = []
        tag_cuisine_and_dish = []
        features_raw = []
        description = []
        ingredients = []
        instructions = []
        restrictions = []
        tag_recipe_keywords = []
        
        url = 'http://receitas.folha.com.br/receita/' + str(i)        
        req = requests.get(url)
        soup = BeautifulSoup(req.text, "lxml")

        try:
            #testa se existe uma receita. Se existe a tag step, então existe
            working = soup.find('div', {'class': 'step'}).text
            print('working -', url)
            
            # Título
            tag_title.append(soup.find('div', {'class': 'title'}).text)

            # tag_cuisine_and_dish
            tag_cuisine_and_dish.append(soup.find('div', {'class': 'cuisine_and_dish'}).text)
            
            # description
            try:
                description.append(soup.find('div', {'itemprop': 'description'}).text)
            except:
                pass

            # download img
#             imgs = soup.findAll("img", {"itemprop":"image"})
#             for img in imgs:
#                 img_url = urljoin(url, img['src'])                
#                 file_name = str(i) + ' - ' + tag_title[0] + '.jpeg'
#                 file_path = os.path.join("img2/", file_name)
#                 try:
#                     urlretrieve(img_url, file_path)
#                 except:
#                     pass
            
            # features_raw = time, "rendimento", 
            for tag in soup.find("div", { "class" : "recipe_image" }).next_siblings:
                if soup.find("div", { "itemprop" : "description" }) == tag:
                    break
                else:
                    features_raw.append(tag)
                    
            # ingredients
            for igr in soup.findAll('span', {'itemprop': 'ingredients'}): 
                ingredients.append(igr.findAll(text = True))

            # instructions
            recipeinstructions = soup.find("div", {"itemprop":"recipeInstructions"}).findAll('p')
            if len(recipeinstructions) >= 1:
                for ins in recipeinstructions:
                    instructions.append(ins.findAll(text = True))
            else:
                recipeinstructions = soup.find("div", {"itemprop":"recipeInstructions"}).findAll('li')
                for ins in recipeinstructions:
                    instructions.append(ins.findAll(text = True))

            # restrictions
            divs = soup.findAll("div", { "class" : "step" })
            for div in divs:
                if div.text=='Restrições Alimentares':
                    restrictions.append(div.next.next.next.text)
                else:pass

            # tag_recipe_keywords
            tag_recipe_keywords.append(soup.find('div', {'class': 'recipe_keywords'}).text)  
            
            # salvando para uma lista para salvar em um df
            lines = [
                i,
                tag_title,
                tag_cuisine_and_dish,
                features_raw,
                description,
                ingredients,
                instructions,
                restrictions,
                tag_recipe_keywords]
            
            # ?
            page.append(lines)
            
        except:
            print('break -',url)
            pass
        
    return page
```

```python
###### executed in 12m 10s, finished 17:23:08 2019-08-01
page = request_receipt()
```

```python
data = pd.DataFrame(page)
data.columns = ['id','receita','tipo','caracteristicas','descricao','ingredientes','instrucoes','restricoes','keyword']
```
